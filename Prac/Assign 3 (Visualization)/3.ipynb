Visualize the data using R/Python by plotting the graphs for assignment no. 1 and 2. Consider suitable data set. Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Heart.csv')
```

```python
df.describe()
```

```python
#type of each column
df.dtypes
```

#pie chart

```python
# plotting pie chart with attributes chestPain, RestBP, Chol, Fbs
#distribution of chestpain using pie chart
df.ChestPain.value_counts().plot.pie(autopct='%.2f')

plt.show()
```

#Bar PLot

Result:
Asymptomatic ChestPain type has highest and typical type of ChestPain is lowest.

```python
# distribution of chestPain type
# using bar plot
sns.countplot(x='ChestPain', data=df)
plt.title('Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
# average cholesterol by AHD
sns.barplot(x='AHD', y='Chol', data=df, estimator=np.mean)
plt.title('Average Cholesterol Levels by AHD Status')
plt.xlabel('AHD (Presence/Absence)')
plt.ylabel('Average Cholesterol')
plt.show()
```

Result:- In the presence of AHD, the cholesterol level is slightly higher

# line chart

```python
# line chart of age and chol
sns.lineplot(x='Age', y='Chol', data=df)
plt.title('Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

Result:- As age increases cholesterol level also increases

#Box Plot


```python
sns.boxplot(x='AHD', y='Chol', data=df)
plt.title('Boxplot of Cholesterol Levels by AHD Status')
plt.xlabel('AHD (Presence/Absence)')
plt.ylabel('Cholesterol')
plt.show()

```

Result:- In the presence of AHD, cholesterol level is slightly higher

```python
sns.boxplot(x='ChestPain', y='Chol', data=df)
plt.title('Boxplot of Cholesterol Levels by ChestPain')
plt.xlabel('ChestPain')
plt.ylabel('Cholesterol')
plt.show()
```

Result:- In asymptomatic ChestPain, cholesterol level is slightly higher

#Scatter Plot


```python
# scatter plot using age and MaxHR
sns.scatterplot(x='Age', y='MaxHR', data=df)
plt.title('Age vs Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.show()
```

Result:- Heart rate is maximum between the age group 50 to 60

```python
# scatter plot using age and RestBP
sns.scatterplot(x='Age', y='RestBP', data=df)
plt.title('Age vs Resting Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.show()
```
